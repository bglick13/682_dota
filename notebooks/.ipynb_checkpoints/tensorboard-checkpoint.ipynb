```python
import torch
import pickle
```

```python
memory_file = 'data/self_play/memories_for_train_3'
mems = []
for mem in os.listdir(memory_file):
    if mem.endswith('pickle'):
        with open(os.path.join(memory_file, mem), 'rb') as f:
            mem = pickle.load(f)
        mems.append(mem)

model: DraftBert = torch.load('weights/final_weights/train_from_selfplay_3.torch')
with open('weights/kmeans.pickle', 'rb') as f:
    clusterizer = pickle.load(f)

dataset = SelfPlayDataset(*mems, clusterizer=clusterizer, le=model.le)
```

```python
make_dot(model(torch.LongTensor(dataset[0][0]).cuda()), params=dict(model.named_parameters()))
```
